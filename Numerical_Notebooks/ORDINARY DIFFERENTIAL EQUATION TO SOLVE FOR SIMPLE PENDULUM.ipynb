```python
"""
ORDINARY DIFFERENTIAL EQUATION TO SOLVE FOR SIMPLE PENDULUM
"""
#importing of libraries such as numpy, scipy.integrate, matplotlib.pyplot, math 
import numpy as np
from scipy.integrate import odeint
import matplotlib.pyplot as plt
import math
#Function that return dz/dt
def model(theta,t,b,g,l,m):
        #defining theta 1 and theta 2
        theta1 = theta[0]
        theta2 = theta[1]
        #defining the dtheta1/dt and dtheta2/dt
        dtheta1_dt = theta2
        dtheta2_dt = -(b/m)*theta2 - (g/l)*math.sin(theta1)
        #defining dtheta/dt
        dtheta_dt = [dtheta1_dt,dtheta2_dt]
        #giving value of dtheta/dt to function
        return dtheta_dt
#Input of the pendulum arrangement 
b = 0.05
g = 9.81
l=1 
m=1
#Damping coefficient
#Acceleration due to gravity
#Length of the string
#Mass of the pendulum
#initial condition [angular_displacement,angular_velocity] 
theta_0 = [0,3]
#Time points starts from 0 to 10 second 
t = np.linspace(0,20,200)
#Solve ODE where model is function,theta_0 is initial input, t is time and args is other argument
theta = odeint(model,theta_0,t,args = (b,g,l,m))
#separating of angular displacement 
angular_displacement= theta[:,0] 
#separating of angular velocity 
angular_velocity = theta[:,1]
#plot result which is stored in figure 1
plt.figure(1)
#olot of t vs angular displacement 
plt.plot(t,theta[:,0],'b-',label=r'$frac{dtheta_1}{dt}=theta2$')
#plot of t vs angular velocity 
plt.plot(t,theta[:,1],'r--',label=r'$frac{dtheta_2}{dt}=-frac{b}{m}theta_2-frac{g} {L}sintheta_1$')
#labelling of plot
plt.ylabel('plot') 
plt.xlabel('time') 
#giving legend to plot 
plt.legend(loc='best') 
#command to show the plot 
plt.show()
#ct is the count time
ct=1
#for loop fo making animation 
for i in angular_displacement:
        #initial condition of string
        x0 = 1
        y0 = 1
        #final condition of the string 
        x1= l-l*math.sin(i)
        y1 = l-l*math.cos(i)
#file name of the image to be saved in the directory 
filename = 'animation of pendulam %03d.png' %ct
#plotting of the figure
plt.figure()
#plot of the wall having linewidth of 5 
plt.plot([0.5,1.5],[1,1],linewidth=5) 
#ploting of the string 
plt.plot([x0,x1],[y0,y1])
#plot of the pendulum 
plt.plot(x1,y1,'o',markersize=15) 
#defining the axis of the plot 
plt.axis([-0.5,2.5,-0.5,1.5])
#saveing of the image in the directory 
plt.savefig(filename)
#increament of the count time
ct= ct+1
```
