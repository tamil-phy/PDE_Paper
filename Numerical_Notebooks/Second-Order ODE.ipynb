```python
import matplotlib.pyplot as plt
import math
from scipy.integrate import odeint
import numpy as np
```

```python
def model(theta,t,b,g,l,m):
    theta1 = theta[0]
    theta2 = theta[1]
    dtheta1_dt = theta2
    dtheta2_dt = -(b/m)*theta2-(g/l)*math.sin(theta1)
    dtheta_dt = [dtheta1_dt,dtheta2_dt]
    return dtheta_dt
# Input parameters
b = 0.05
g = 9.81
l = 1
m = 1
# Initial Condition
theta_0 = [0,3]
t = np.linspace(0,20,250) #time point
theta = odeint(model,theta_0,t,args=(b,g,l,m))
```

```python
#olot of t vs angular displacement 
plt.plot(t,theta[:,0],'b-',label=r'$frac{dtheta_1}{dt}=theta2$')
#plot of t vs angular velocity 
plt.plot(t,theta[:,1],'r--',label=r'$frac{dtheta_2}{dt}=-frac{b}{m}theta_2-frac{g} {L}sintheta_1$')
#labelling of plot
plt.ylabel('plot') 
plt.xlabel('time') 
#giving legend to plot 
plt.legend(loc='best') 
#command to show the plot 
plt.show()
```
